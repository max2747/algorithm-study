```python
def solution(p):
    def right_p(s):
        temp = []
        if s[0] == ')' : return False
        for idx, val in enumerate(s):
            if val == '(' : temp.append(val)
            else:
                if len(temp) == 0: return False
                else: temp.pop()
        if len(temp) == 0: return True
        else: return False
    
    def balance_idx(s):
        count = 0
        for i in range(len(s)):
            if s[i] == '(': count += 1
            if s[i] == ')': count -= 1
            if i > 0 and count == 0: return i+1
        
    
    answer = ''
    if p == '' or right_p(p) : return p
    u, v = p[:balance_idx(p)], p[balance_idx(p):]
    
    if right_p(u):
        temp = solution(v)
        answer = u + temp
    else:
        temp = '(' + solution(v) + ')'
        temp += "".join(['(' if i == ')' else ')' for i in u[1:-1]  ])
        answer += temp
    return answer
```
